```python
import pandas as pd

# Updated URL to access raw CSV data
url = 'https://raw.githubusercontent.com/prvpitawala/ML-Project-1-Pre-Process-Simple-Dataset-with-Basic-Pre-Processing-Techniques/main/Data.csv'
df = pd.read_csv(url)
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
for col in df.columns:
    if df[col].dtype == 'int64' or df[col].dtype == 'float64':
        df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
```

```python
df.to_csv("Tidy_Dataset.csv", index=False)
```
